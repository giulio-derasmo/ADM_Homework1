```python
#sWAP cASE

def swap_case(s):
    strg = ""        #initialize string
    for char in s:
        if char.isupper():
            char = char.lower()
            strg += char            #recostruction of the string
        else:
            char = char.upper()
            strg += char
    return strg

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
#Text Wrap

def wrap(string, max_width):
    line = [char for char in string]
    for i in range(max_width,len(line)+len(line)//max_width,max_width+1):
        line.insert(i,"\n")
    
    string = "".join(line)
    return string

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```

```python
#Capitalize!

def solve(s):
    s = s.lower()
    string = s[0].upper()
    char_old = s[0]
    for char in s[1:]:
        if char_old.isspace():
            string += char.upper()
        else: 
            string += char
        char_old = char
    return string

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = solve(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#The Minion Game

def minion_game(string):
    # your code goes here
    S, K = 0,0   #partono con 0 punti
    vocali = "AEIOU"
    for i in range(len(string)):
        if string[i] in vocali:
            K = K + len(string) - i
        else:
            S = S + len(string) - i    
    
    if S>K:
        print("Stuart",S)
    elif S==K:
        print("Draw")    
    else:
        print("Kevin",K)  
    

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
#Collections.deque()

d = deque()
N = int(input())

for i in range(N):
    inp = input().split()
    if inp[0] == "append":
        d.append(inp[-1])
    elif inp[0] == "appendleft":
        d.appendleft(inp[-1])
    elif inp[0] == "pop":
        d.pop()
    else:
        d.popleft()

print(" ".join(list(d)))
```

```python
#Piling Up!

T = int(input())

for i in range(T):
    n = int(input())
    d = deque(list(map(int, input().split())))
    tmp = -1
    for w in range(n):
        left, right = d[0], d[-1]
        if left >= right:
            if tmp >= left or tmp == -1:
                tmp = left
                d.popleft()
        else:
            if tmp >= right or tmp == -1:
                tmp = right
                d.pop()
            
    if len(d) == 0:
        print("Yes")
    else:
        print("No")
            
```

```python
#Time Delta

def time_delta(t1, t2):
    formato = '%a %d %b %Y %H:%M:%S %z'
    t1 = datetime.strptime(t1, formato)    #crea datetime.object 
    t2 = datetime.strptime(t2, formato)
    return str(int(abs((t1-t2).total_seconds())))  

    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
```

```python
#Detect Floating Point Number

#comincia con + o - 1 volta
#poi tutti numeri, poi punto e numeri
#* perche posso avere .numero
#+ perche dopo . ci deve essere numero

T = int(input())
pattern = r"^([+-]?)\d*\.\d+$"
for i in range(T):
    number = input()
    match = re.match(pattern,number)
    if match == None:
        print(False)
    else:
        print(True)
```

```python
#HTML Parser - Part 1

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Start :", tag)
        for att in attrs:
            print("->", att[0], ">", att[1])
    def handle_endtag(self, tag):
        print("End   :", tag)
    def handle_startendtag(self, tag, attrs):
        print("Empty :", tag)
        for att in attrs:
            print("->", att[0], ">", att[1])       
   
            
N = int(input())
parser = MyHTMLParser()

for i in range(N):
    parser.feed(input())
```

```python
#HTML Parser - Part 2

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if "\n" in data:     #print(data) staccato per motivi di indentatura
            print(">>> Multi-line Comment")
        else:
            print(">>> Single-line Comment")
        print(data)
    def handle_data(self, data):
        if data == '\n':
            return               #non print quando data == \n
        print(">>> Data")
        print(data)
  
  
html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#Validating Credit Card Numbers

#pattern1 = cheack se: 16 digits, inizia con 4,5o 6
# e se ci sta - tra 4 numeri
#pattern2 = cheack se i numeri si ripetono piu di 3

pattern1 = r"^(4|5|6)(\d){3}(\-)?(\d){4}(\-)?(\d){4}(\-)?(\d){4}$"   
pattern2 = r"(\d)\1{3,}"

N = int(input())
for i in range(N):
    number = input()
    match = re.match(pattern1,number)
    if match: #tolgo il -
        number = ''.join(c for c in number if c.isdigit())
        match2 = re.search(pattern2,number)
        if not match2:
            print("Valid")
        else:
            print("Invalid")
    else: 
        print("Invalid")
        
#re.search per cercare nella riga, oppure re.match ma con pattern2 = r".*(\d)\1{3,}"
```

```python
#Validating Postal Codes

#^[1-9]\d{5}$ inizia con num tra 1-9 poi finisce con 5 digits
#matcha se \d è seguito da un numero e la ripetizione del gruppo 1 ossia (\d) 
regex_integer_in_range = r"^[1-9]\d{5}$" # Do not delete 'r'.
regex_alternating_repetitive_digit_pair = r"(\d)(?=\d\1)" # Do not delete 'r'

P = input()

print (bool(re.match(regex_integer_in_range, P)) 
and len(re.findall(regex_alternating_repetitive_digit_pair, P)) < 2)
```

```python
#Regex Substitution

N = int(input())
#match se prima e dopo || o && ci sta uno spazio
pattern = r"(?<= )((\|\|)|(\&\&))(?= )"

def replace(match):
    if match.group(0) == "&&":
        return "and"
    else:
        return "or"
    
for i in range(N):
    string = input()
    print(re.sub(pattern,replace, string))
    

```

```python
#Validating UID

T = int(input())
#pattern1 = 
#controllo A-Z at least 2 + [0-9] at least 3 + len(string) == 10
#pattern2 =
#match se due caratteri si ripetono non consecutivamente
#pattern3 = 
#match se due caratteri si ripetono consecutivamente

pattern = r"(?=.*((.*[A-Z]){2,}))(?=.*((.*[0-9]){3,}))(?=^(([a-zA-Z0-9]){10})$)"
pattern2 = r"([a-zA-Z0-9])(?=[a-zA-Z0-9].*\1)"
pattern3 = r"([a-zA-Z0-9])\1"  #repetion
for i in range(T):
    string = input()
    if string.isalnum():
        if bool(re.search(pattern,string)) and not bool(re.search(pattern2,string)) and not bool(re.search(pattern3,string)):
            print("Valid")
        else:
            print("Invalid")
    else:
        print("Invalid")
```

```python
#Matrix Script

first_multiple_input = input().rstrip().split()

n = int(first_multiple_input[0])

m = int(first_multiple_input[1])

matrix = []

for _ in range(n):
    matrix_item = input()
    matrix.append(matrix_item)


matrix = list(zip(*matrix))  #reshape matrix  +  lista di tuple
matrix = [item for el in matrix for item in el]  #lista
string = "".join(matrix)  #stringa

#(?<=---) x (?=---) controllo prima e dopo x
#[^x] match any character except x (\w è [a-zA-Z0-9], ma scritto [] non lo prende
#+ perche ce ne possono essere 0 o piu
pattern = r"(?<=([a-zA-Z0-9]))([^\w)])+(?=([a-zA-Z0-9]))"
print(re.sub(pattern," ",string))
```

# Closures and Decorations

```python
#Standardize Mobile Number Using Decorators

def wrapper(f):
    def fun(l):
        # complete the function
        f(["+91 " + c[-10:-5] + " " + c[-5:] for c in l])
    return fun

@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 
```

```python
#Decorators 2 - Name Directory

def person_lister(f):
    def inner(people):
        # complete the function
        return [ f(p) for p in sorted(people, key = lambda x: (int(x[2])))]
    return inner

@person_lister
def name_format(person):
    return ("Mr. " if person[3] == "M" else "Ms. ") + person[0] + " " + person[1]

if __name__ == '__main__':
    people = [input().split() for i in range(int(input()))]
    print(*name_format(people), sep='\n')
```
